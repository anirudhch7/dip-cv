# Color Channel Swapping
This section demonstrates how to manipulate image color channels by swapping the Red and Blue channels using NumPy.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os

# Enhanced: Color Channel Swap with Error Handling
try:
    # Load image with validation
    if not os.path.exists("lenna.png"):
        raise FileNotFoundError("Image 'lenna.png' not found")

    image = mpimg.imread("lenna.png")

    # Extract individual color channels
    r = image[:,:,0]
    g = image[:,:,1]
    b = image[:,:,2]

    # Create swapped image (swap red and blue channels)
    swap_image = np.copy(image)
    swap_image[:,:,0] = b  # Blue -> Red
    swap_image[:,:,2] = r  # Red -> Blue

    # Enhanced visualization with better layout
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].imshow(image)
    axes[0].set_title("Original Image (RGB)", fontsize=12, fontweight='bold')
    axes[0].axis('off')

    axes[1].imshow(swap_image)
    axes[1].set_title("Channel Swapped (BGR)", fontsize=12, fontweight='bold')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")
```

# Bitwise Operations
This section covers fundamental bitwise operations (AND, OR, XOR, NOT) used in image masking and shape manipulation.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

try:
    # Create two simple images
    image1 = np.zeros((300, 300))
    cv2.rectangle(image1, (50, 50), (250, 250), 255, -1)  # White rectangle

    image2 = np.zeros((300, 300))
    cv2.circle(image2, (150, 150), 100, 255, -1)  # White circle

    # Bitwise AND operation
    bitwise_and = cv2.bitwise_and(image1, image2)

    # Bitwise OR operation
    bitwise_or = cv2.bitwise_or(image1, image2)

    # Bitwise XOR operation
    bitwise_xor = cv2.bitwise_xor(image1, image2)

    # Bitwise NOT operation
    bitwise_not1 = cv2.bitwise_not(image1)
    bitwise_not2 = cv2.bitwise_not(image2)

    # Enhanced visualization with subplots
    fig, axes = plt.subplots(2, 4, figsize=(16, 8))

    results = [
        (image1, 'Image 1 (Rectangle)', axes[0, 0]),
        (image2, 'Image 2 (Circle)', axes[0, 1]),
        (bitwise_and, 'Bitwise AND', axes[0, 2]),
        (bitwise_or, 'Bitwise OR', axes[0, 3]),
        (bitwise_xor, 'Bitwise XOR', axes[1, 0]),
        (bitwise_not1, 'NOT Image 1', axes[1, 1]),
        (bitwise_not2, 'NOT Image 2', axes[1, 2]),
    ]

    for image, title, ax in results:
        ax.imshow(image, cmap='gray')
        ax.set_title(title, fontweight='bold')
        ax.axis('off')

    axes[1, 3].axis('off')  # Hide the last empty subplot
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error in bitwise operations: {e}")
```

# Line Detection with Hough Transform
This section uses Canny edge detection and the Probabilistic Hough Line Transform to identify and draw lines in an image.

```python
import cv2
import numpy as np
import os

try:
    # Validate image file exists
    if not os.path.exists('lines.jpg'):
        raise FileNotFoundError("Image 'lines.jpg' not found")

    # Read image and convert to grayscale
    image = cv2.imread('lines.jpg')
    if image is None:
        raise ValueError("Failed to load image")

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Canny edge detection with tunable parameters
    canny_low = 50
    canny_high = 200
    edges = cv2.Canny(gray, canny_low, canny_high)

    # Detect lines using HoughLinesP
    lines = cv2.HoughLinesP(
        edges,
        rho=1,
        theta=np.pi / 180,
        threshold=68,
        minLineLength=15,
        maxLineGap=200
    )

    # Draw detected lines
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(image, (x1, y1), (x2, y2), (255, 255, 0), 2)  # Cyan color, thicker line
        print(f"Detected {len(lines)} lines")
    else:
        print("No lines detected")

    # Display results
    cv2.imshow('Original', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error processing image: {e}")
```

# Image Filtering and Thresholding
This section demonstrates various spatial filtering techniques (Mean, Median, Gaussian) and Otsu's automated thresholding method.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

try:
    # Validate image exists
    if not os.path.exists("lenna.png"):
        raise FileNotFoundError("Image 'lenna.png' not found")

    # Load the image
    original_image = cv2.imread("lenna.png", cv2.IMREAD_GRAYSCALE)

    if original_image is None:
        raise ValueError("Failed to load image")

    # Apply mean filter
    mean_filtered = cv2.blur(original_image, (5, 5))

    # Apply median filter
    median_filtered = cv2.medianBlur(original_image, 5)

    # Apply Gaussian filter
    gaussian_filtered = cv2.GaussianBlur(original_image, (5, 5), 0)

    # Apply Otsu Thresholding
    _, otsu_thresholded = cv2.threshold(original_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Display original and processed images with histograms
    fig, axes = plt.subplots(5, 2, figsize=(12, 16))

    # Define images and titles
    images = [
        (original_image, 'Original Image'),
        (mean_filtered, 'Mean Filtered'),
        (median_filtered, 'Median Filtered'),
        (gaussian_filtered, 'Gaussian Filtered'),
        (otsu_thresholded, 'Otsu Thresholded'),
    ]

    for idx, (img, title) in enumerate(images):
        # Display image
        axes[idx, 0].imshow(img, cmap='gray')
        axes[idx, 0].set_title(title, fontweight='bold')
        axes[idx, 0].axis('off')

        # Display histogram
        hist = cv2.calcHist([img], [0], None, [256], [0, 256])
        axes[idx, 1].plot(hist, color='blue')
        axes[idx, 1].set_title(f'{title} Histogram')
        axes[idx, 1].set_xlim([0, 256])
        axes[idx, 1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("Image filtering and histogram analysis completed successfully!")

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error in filtering operations: {e}")
```

# Edge Detection Techniques
Comparing Canny and Sobel operators for detecting intensity discontinuities (edges) in an image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

try:
    # Validate image exists
    if not os.path.exists('lenna.png'):
        raise FileNotFoundError("Image 'lenna.png' not found")

    # Load the binary image
    binary_image = cv2.imread('lenna.png', cv2.IMREAD_GRAYSCALE)

    if binary_image is None:
        raise ValueError("Failed to load image")

    # Apply Canny filter for edge detection
    canny_edges = cv2.Canny(binary_image, 90, 100)

    # Apply Sobel filter for edge detection
    sobel_x = cv2.Sobel(binary_image, cv2.CV_64F, 0, 1, ksize=3)
    sobel_y = cv2.Sobel(binary_image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_edges = cv2.magnitude(sobel_x, sobel_y)

    # Normalize Sobel edges to 0-255 range for better visualization
    sobel_edges = np.uint8(255 * sobel_edges / np.max(sobel_edges))

    # Display the results
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    images = [
        (binary_image, 'Original Image'),
        (canny_edges, 'Canny Edges'),
        (sobel_edges, 'Sobel Edges')
    ]

    for ax, (img, title) in zip(axes, images):
        ax.imshow(img, cmap='gray')
        ax.set_title(title, fontweight='bold')
        ax.axis('off')

    plt.tight_layout()
    plt.show()

    print("Edge detection completed successfully!")

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error in edge detection: {e}")
```

# Haar Cascade Face Detection
Using pre-trained Haar Cascade classifiers to detect human faces within an image.

```python
import cv2
import os

try:
    # Validate image exists
    if not os.path.exists("lenna.png"):
        raise FileNotFoundError("Image 'lenna.png' not found")

    # Load the pre-trained Haar Cascade classifier for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    img = cv2.imread("lenna.png")

    if img is None:
        raise ValueError("Failed to load image")

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Perform face detection with tunable parameters
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))

    # Draw rectangles around the detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Print detection results
    print(f"Number of faces detected: {len(faces)}")
    for idx, (x, y, w, h) in enumerate(faces):
        print(f"Face {idx + 1}: Position ({x}, {y}), Size ({w}x{h})")

    # Display the result
    cv2.imshow('Face Detection', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

except FileNotFoundError as e:
    print(f"Error: {e}")
except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error in face detection: {e}")
```

# Morphological Operations
Demonstrating Erosion, Dilation, Opening, Closing, and Gradient operations using a structuring element.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

try:
    # Validate image exists
    if not os.path.exists("M.jpg"):
        raise FileNotFoundError("Image 'M.jpg' not found")

    # Load image
    img = cv2.imread("M.jpg", cv2.IMREAD_GRAYSCALE)

    if img is None:
        raise ValueError("Failed to load image")

    # Define morphological kernel
    kernel = np.ones((3, 3), np.uint8)

    # Apply morphological operations
    eroded = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)
    dilated = cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel)
    opened = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    closed = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

    # Create visualization
    fig, axes = plt.subplots(2, 3, figsize=(14, 8))

    operations = [
        (img, "Original Image"),
        (eroded, "Eroded"),
        (dilated, "Dilated"),
        (opened, "Opened"),
        (closed, "Closed"),
        (gradient, "Gradient"),
    ]

    for ax, (image, title) in zip(axes.flat, operations):
        ax.imshow(image, cmap='gray')
        ax.set_title(title, fontweight='bold')
        ax.axis('off')

    plt.tight_layout()
    plt.show()

    print("Morphological operations completed successfully!")

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error in morphological operations: {e}")
```

# Histogram Equalization and CLAHE
Enhancing image contrast using global Histogram Equalization and Contrast Limited Adaptive Histogram Equalization (CLAHE).

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

try:
    # Validate image exists
    if not os.path.exists("lenna.png"):
        raise FileNotFoundError("Image 'lenna.png' not found")

    # Load image
    img = cv2.imread("lenna.png", cv2.IMREAD_GRAYSCALE)

    if img is None:
        raise ValueError("Failed to load image")

    # Standard histogram equalization
    eimg = cv2.equalizeHist(img)

    # Enhanced: CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    clahe_img = clahe.apply(img)

    # Calculate histograms
    hi = cv2.calcHist(img, [0], None, [256], [0, 256])
    he = cv2.calcHist(eimg, [0], None, [256], [0, 256])
    hc = cv2.calcHist(clahe_img, [0], None, [256], [0, 256])

    # Create enhanced visualization
    fig = plt.figure(figsize=(14, 10))

    # Original image and histogram
    plt.subplot(3, 2, 1)
    plt.title("Original Image", fontweight='bold')
    plt.imshow(img, cmap="gray")
    plt.axis('off')

    plt.subplot(3, 2, 2)
    plt.title("Original Histogram", fontweight='bold')
    plt.plot(hi, color='blue')
    plt.xlim([0, 256])
    plt.grid(True, alpha=0.3)

    # Standard equalization
    plt.subplot(3, 2, 3)
    plt.title("Histogram Equalized", fontweight='bold')
    plt.imshow(eimg, cmap="gray")
    plt.axis('off')

    plt.subplot(3, 2, 4)
    plt.title("Equalized Histogram", fontweight='bold')
    plt.plot(he, color='green')
    plt.xlim([0, 256])
    plt.grid(True, alpha=0.3)

    # CLAHE enhancement
    plt.subplot(3, 2, 5)
    plt.title("CLAHE Enhanced", fontweight='bold')
    plt.imshow(clahe_img, cmap="gray")
    plt.axis('off')

    plt.subplot(3, 2, 6)
    plt.title("CLAHE Histogram", fontweight='bold')
    plt.plot(hc, color='red')
    plt.xlim([0, 256])
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("Histogram equalization completed successfully!")

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error in histogram equalization: {e}")
```

# Frequency Domain Filtering
Using the Fast Fourier Transform (FFT) to apply Low-Pass and High-Pass filters in the frequency domain.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

try:
    # Validate image exists
    if not os.path.exists('lenna.png'):
        raise FileNotFoundError("Image 'lenna.png' not found")

    # Read the input image
    img = cv2.imread('lenna.png', cv2.IMREAD_GRAYSCALE)

    if img is None:
        raise ValueError("Failed to load image")

    # Get image dimensions
    rows, cols = img.shape
    center_row, center_col = rows // 2, cols // 2

    # Create low-pass and high-pass filter masks
    lpf_mask = np.zeros((rows, cols), np.uint8)
    lpf_mask[center_row-30:center_row+30, center_col-30:center_col+30] = 1

    hpf_mask = 1 - lpf_mask

    # Perform Fourier Transform on the image
    f_transform = np.fft.fft2(img)
    f_transform_shifted = np.fft.fftshift(f_transform)

    # Calculate magnitude spectrum for visualization
    magnitude_spectrum = 20 * np.log(np.abs(f_transform_shifted) + 1)

    # Apply the filters to the frequency domain image
    f_transform_shifted_lpf = f_transform_shifted * lpf_mask
    f_transform_shifted_hpf = f_transform_shifted * hpf_mask

    # Inverse Fourier Transform to get the enhanced images
    img_enhanced_lpf = np.abs(np.fft.ifft2(np.fft.ifftshift(f_transform_shifted_lpf)))
    img_enhanced_hpf = np.abs(np.fft.ifft2(np.fft.ifftshift(f_transform_shifted_hpf)))

    # Normalize for better visualization
    img_enhanced_lpf = np.uint8(255 * img_enhanced_lpf / np.max(img_enhanced_lpf))
    img_enhanced_hpf = np.uint8(255 * img_enhanced_hpf / np.max(img_enhanced_hpf))

    # Display results with frequency domain visualization
    fig, axes = plt.subplots(2, 3, figsize=(16, 10))

    # Row 1: Spatial domain
    axes[0, 0].imshow(img, cmap='gray')
    axes[0, 0].set_title('Original Image', fontweight='bold')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(magnitude_spectrum, cmap='hot')
    axes[0, 1].set_title('Magnitude Spectrum', fontweight='bold')
    axes[0, 1].axis('off')

    axes[0, 2].imshow(img, cmap='gray')
    axes[0, 2].set_title('Original (for reference)', fontweight='bold')
    axes[0, 2].axis('off')

    # Row 2: Filtered results
    axes[1, 0].imshow(img_enhanced_lpf, cmap='gray')
    axes[1, 0].set_title('Low-Pass Filtered', fontweight='bold')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(magnitude_spectrum, cmap='hot')
    axes[1, 1].set_title('LPF Mask Applied', fontweight='bold')
    axes[1, 1].axis('off')

    axes[1, 2].imshow(img_enhanced_hpf, cmap='gray')
    axes[1, 2].set_title('High-Pass Filtered', fontweight='bold')
    axes[1, 2].axis('off')

    plt.tight_layout()
    plt.show()

    print("Frequency domain filtering completed successfully!")

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Error in frequency domain filtering: {e}")
```
